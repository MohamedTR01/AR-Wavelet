```python
# pip install PyWavelets

import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pywt # 'pywt' signifie PyWavelets, qui est une bibliothque utilise pour les transformations d'ondelettes en Python
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import adfuller
import datetime
from statsmodels.tsa.ar_model import AutoReg

# Spcifiez le symbole de l'action que vous souhaitez importer
symbol = "AAPL"

# Utilisez la fonction `download` de yfinance pour obtenir les donnes
data = yf.download(symbol, start="2011-01-01", end="2023-06-20")

# Slectionnez uniquement la colonne "Adj Close"
adj_close = data["Adj Close"]
print(adj_close)

# Crez un graphique avec les donnes "Adj Close"
plt.figure(figsize=(10, 6))
plt.plot(adj_close)
plt.title("Prix ajust en clture AAPL")
plt.xlabel("Date")
plt.ylabel("Prix ajust en clture")
plt.grid(True)
plt.show()
```

```python
# Convertir les donnes en une liste
signal = adj_close.values.tolist()

# Effectuer la dcomposition en ondelettes
coeffs = pywt.wavedec(signal, 'db4', level=4)

# Rcuprer les coefficients de dtail et l'approximation
approximation = coeffs[0]
details = coeffs[1:]

# Afficher les coefficients de dtail
plt.figure(figsize=(10, 6))
for i, detail_level in enumerate(details):
    plt.subplot(len(details), 1, i+1)
    plt.plot(detail_level)
    plt.title(f"Detail Level {i+1}")
plt.tight_layout()
plt.show()
```

```python
# Convertir les donnes en une liste
signal = adj_close.values.tolist()

# Effectuer la dcomposition en ondelettes
coeffs = pywt.wavedec(signal, 'db4', level=4)

# Prvision des valeurs futures  chaque niveau de dtail
for i, detail_level in enumerate(coeffs[1:]):
    # Cration du modle autorgressif (AR) avec un ordre (lag) de 10
    model = AutoReg(detail_level, lags=10)

    # Ajustement du modle aux donnes
    model_fit = model.fit()

    # Prvision des valeurs futures  partir du modle ajust
    predictions = model_fit.predict(start=len(detail_level), end=len(detail_level) + 10)

    # Affichage des prdictions
    plt.figure(figsize=(10, 6))
    plt.plot(detail_level, label="Donnes historiques")
    plt.plot(range(len(detail_level), len(detail_level) + len(predictions)), predictions, label="Prvisions")
    plt.title(f"Prvisions - Niveau de dtail {i+1}")
    plt.xlabel("Time")
    plt.ylabel("Signal")
    plt.legend()
    plt.show()
```

```python
# Spcifiez le symbole de l'action que vous souhaitez importer
symbol = "AAPL"

# Utilisez la fonction `download` de yfinance pour obtenir les donnes
data = yf.download(symbol, start="2011-01-01", end="2023-06-20")

# Slectionnez uniquement la colonne "Adj Close"
adj_close = data["Adj Close"]

# Convertir les donnes en une liste
signal = adj_close.values.tolist()

# Effectuer la dcomposition en ondelettes
coeffs = pywt.wavedec(signal, 'db4', level=4)

# Analyse de la stationarit pour chaque niveau de dtail
for i, detail_level in enumerate(coeffs[1:]):
    # Effectuer le test de Dickey-Fuller augment
    result = adfuller(detail_level)

    # Extraire la p-value
    p_value = result[1]

    # Interprtation de la stationnarit
    if p_value < 0.05:
        stationarity = "Stationnaire"
    else:
        stationarity = "Non stationnaire"

    # Afficher le rsultat
    print(f"Niveau de dtail {i+1} : {stationarity} (p-value: {p_value})")
```

```python
# Spcifiez le symbole de l'action que vous souhaitez importer
symbol = "AAPL"

# Utilisez la fonction `download` de yfinance pour obtenir les donnes
data = yf.download(symbol, start="2011-01-01", end="2023-06-20")

# Slectionnez uniquement la colonne "Adj Close"
adj_close = data["Adj Close"]

# Convertir les donnes en une liste
signal = adj_close.values.tolist()

# Effectuer la dcomposition en ondelettes
coeffs = pywt.wavedec(signal, 'db4', level=4)

# Prvision des valeurs futures  chaque niveau de dtail
predictions_combined = []
for i, detail_level in enumerate(coeffs[1:]):
    # Cration du modle autorgressif (AR) avec un ordre (lag) de 10
    model = LinearRegression()

    # Ajustement du modle aux donnes
    model.fit(np.arange(len(detail_level)).reshape(-1, 1), detail_level)

    # Prvision des valeurs futures  partir du modle ajust
    predictions = model.predict(np.arange(len(detail_level), len(detail_level) + 10).reshape(-1, 1))

    # Ajouter les prdictions au signal combin
    predictions_combined.append(predictions)

# Transformer les prvisions combines en un tableau 2D
predictions_combined = np.array(predictions_combined).T

# Rpter les prvisions pour correspondre  la longueur du signal d'origine
repeated_predictions = np.repeat(predictions_combined, len(signal) // len(predictions_combined) + 1, axis=0)
repeated_predictions = repeated_predictions[:len(signal)]

# Rgression linaire pour estimer le signal d'origine
regressor = LinearRegression()
regressor.fit(repeated_predictions, signal)

# Prvision finale du signal d'origine
original_signal_predictions = regressor.predict(repeated_predictions)

# Prvision pour l'anne 2024
year_2024_predictions = original_signal_predictions[-252:]

# Afficher les prvisions pour l'anne 2024
plt.figure(figsize=(10, 6))
plt.plot(signal, label="Donnes historiques")
plt.plot(range(len(signal), len(signal) + len(year_2024_predictions)), year_2024_predictions, label="Prvisions 2024")
plt.title("Prvisions - Signal d'origine pour 2024")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.legend()
plt.show()
```

```python
# Spcifiez le symbole de l'action que vous souhaitez importer
symbol = "AAPL"

# Utilisez la fonction `download` de yfinance pour obtenir les donnes
data = yf.download(symbol, start="2011-01-01", end=datetime.datetime.now())

# Slectionnez uniquement la colonne "Adj Close"
adj_close = data["Adj Close"]

# Convertir les donnes en une liste
signal = adj_close.values.tolist()

# Effectuer la dcomposition en ondelettes
coeffs = pywt.wavedec(signal, 'db4', level=4)

# Liste pour stocker les prvisions de chaque niveau de dtail
predictions_combined = []

# Prvision des valeurs futures  chaque niveau de dtail
for i, detail_level in enumerate(coeffs[1:]):
    # Cration du modle autorgressif (AR) avec un ordre (lag) de 10
    model = AutoReg(detail_level, lags=10)

    # Ajustement du modle aux donnes
    model_fit = model.fit()

    # Prvision des valeurs futures  partir du modle ajust
    predictions = model_fit.predict(start=len(detail_level), end=len(detail_level) + 10)

    # Ajouter les prdictions au signal combin
    predictions_combined.append(predictions)

# Transformer les prvisions combines en un tableau 2D
predictions_combined = np.array(predictions_combined)

# Calculer la moyenne des prvisions pour chaque priode de temps
average_predictions = np.mean(predictions_combined, axis=0)

# Rpter les prvisions pour correspondre  la longueur du signal d'origine
repeated_predictions = np.repeat(average_predictions, len(signal) // len(average_predictions) + 1, axis=0)
repeated_predictions = repeated_predictions[:len(signal)]

# Afficher les prvisions moyennes
plt.figure(figsize=(10, 6))
plt.plot(signal, label="Donnes historiques")
plt.plot(range(len(signal), len(signal) + len(repeated_predictions)), repeated_predictions, label="Prvisions moyennes")
plt.title("Prvisions - Signal d'origine pour les 10 prochaines annes")
plt.xlabel("Time")
plt.ylabel("Signal")
plt.legend()
plt.show()
```
